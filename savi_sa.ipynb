# seaborn


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

[link text](https://)## Univariate → “one variable” data visualization

```python
#strip plot
sns.stripplot(x='sex',y='tip', data=tips,jitter=False)
plt.title('Strip plot')
plt.xlabel('Gender')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='sex', y='tip',hue='day', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot
plt.figure(figsize=(10,5))

sns.swarmplot(x='day', y='tip', data=tips,facecolor='black')
sns.violinplot(x='day',y='tip',data=tips,color='red')

plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

## Box and Violin plot

```python
#plt.subplot(1,2,1)
plt.figure(figsize=(10,10))
sns.violinplot(x='day', y='tip',data=tips)
sns.boxplot(x='day', y='tip', data=tips)

plt.ylabel('tip ($)')
plt.show()
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
# subplots
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
##Combining plots

sns.violinplot(x='day', y='tip', data=tips, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

## Bivariate → “two variables” data visualization

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
# Contours are created representing various levels of density of the data
```

```python
pd.plotting.scatter_matrix(tips)
plt.show()
```

```python
#Pair plot
sns.pairplot(tips)
plt.show()
```

```python
#Pair plot
sns.pairplot(tips)
plt.show()
```
